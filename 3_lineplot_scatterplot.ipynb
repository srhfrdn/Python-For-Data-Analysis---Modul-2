# __1. Line Plot / Line Chart__

__Wikipedia__

*A line chart or line plot or line graph or curve chart is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments.  A line chart is often used to visualize a trend in data over intervals of time – a time series – thus the line is often drawn chronologically. In these cases they are known as run charts.*

## __When to use line charts__
A line chart allows us to track the development of several variables at the same time. It is very easy to understand, and the reader doesn’t feel overwhelmed.

### Time Series by line plot
Time series is a line plot and it is basically connecting data points with a straight line. It is useful in understanding the trend over time. It can explain the correlation between points by the trend. An upward trend means positive correlation and downward trend means a negative correlation. It mostly used in forecasting, monitoring models.

__When to use__: Time Series should be used when single or multiple variables are to be plotted over time.

eg: Stock Market Analysis of Companies, Weather Forecasting.

# __2. Scatter Plot__

A scatter plot is a type of chart that is often used in the fields of statistics and data science. If a scatter plot includes more than two variables, then we would use different colours to signify that.

Scatter plot helps in visualizing 2 numeric variables. It helps in identifying the relationship of the data with each variable i.e correlation or trend patterns. It also helps in detecting outliers in the plot.

__When to use__: ``A scatter plot chart is a great indicator that allows us to see whether there is a pattern to be found between two variables.``

It is used in Machine learning concepts like regression, where x and y are continuous variables. It is also used in clustering scatters or outlier detection.

<img src='https://365datascience.com/wp-content/uploads/2018/07/25-Scatter-plot-with-correlation-2.png' widht=100 height=100\>

<img src='https://365datascience.com/wp-content/uploads/2018/07/26-scatter-plot-no-relationship-2.png' widht=100 height=100\>

*So, this can be a very useful chart type whenever we would like to see if there is any relationship between two sets of data.*


``plt.scatter()`` takes 2 numeric arguments for scattering data points in the plot. It is similar to line plot except without the connected straight lines. By corr we mean correlation 

### __When to avoid scatter plots__

We can’t use scatter plots when we don’t have bi-dimensional data. 
In our example, we need information about both house prices and house size to create a scatter plot. 
A scatter plot requires at least two dimensions for our data.

In addition, scatter plots are not suitable if we are interested in observing time patterns.

Finally, a scatter plot is used with numerical data, or numbers. If we have categories such as 3 divisions, 5 products, and so on, a scatter plot would not reveal much.

# __STACK PLOT__

```python
## kayak piechart yg punya dimensi waktu
#cocok pake data yg fix ( yaitu ada 0-100%)
#lebih baik convert dulu datanya jadi persentase buat stacked area 100%chart

#chart dimana jumlah akhirnya sama
```

```python
##membuat stackplot tentang kegiatan dalam 24jam

#bisa dijadikan value, karena data atas dan bawah sudah fix yaitu 0 dan 24

days = [1,2,3,4,5]

sleeping = [7,4,5,6,3]
eating = [3,4,5,6,3]
working = [8,8,8,8,8]
playing = [6,8,6,4,10]
                
            # x   ,  y    
plt.stackplot(days, [sleeping,eating,working,playing], labels=('sleeping', 'eating','working','playing'))
plt.xlabel('Days')
plt.ylabel('Hour Activity')
plt.ylim(0,24) ##ylim = mengatur sumbu y (awal, akhir)
plt.legend()
plt.show()



plt.show()
```

```python
sleeping_perc =pd.Series(sleeping).divide(24)
eating_perc=pd.Series(eating).divide(24)
working_perc=pd.Series(working).divide(24)
playing_perc=pd.Series(playing).divide(24)
```

```python
#sama kyk yg diatas tapi dijadikan rasio

activity = np.vstack([sleeping_perc, eating_perc, working_perc, playing_perc])
plt.stackplot(days, activity, labels=('sleeping', 'eating', 'working', 'playing'))
plt.xlabel('Days')
plt.ylabel('Activity')
# plt.ylim(0,24)
plt.legend()
plt.show()
```

```python
#membuat stackplot volume transaksi saham dari 4 emiten
```

```python
#vstack data volume
volume = np.vstack([bbri['Volume'], icbp['Volume'], sido['Volume'], bmri['Volume']])

#dibikin jadi dataframe
vol_df = pd.DataFrame(volume, index=['BBRI', 'ICBP', 'SIDO', 'BMRI'], columns=bbri['Date'])

#lalu di transpose
vol_df = vol_df.T
vol_df
```

```python
#membuat kolom jumlah

#axis = 1 ==> totalin per baris
vol_df['Total'] = vol_df[['BBRI', 'ICBP', 'SIDO', 'BMRI']].sum(axis=1)

#membuat percentage
vol_df_perc = vol_df.copy()
vol_df_perc
```

```python
vol_df_perc['BBRI']= vol_df['BBRI']/vol_df['Total']*100
vol_df_perc['ICBP']= vol_df['ICBP']/vol_df['Total']*100
vol_df_perc['SIDO']= vol_df['SIDO']/vol_df['Total']*100
vol_df_perc['BMRI']= vol_df['BMRI']/vol_df['Total']*100

vol_df_perc['Total'] = vol_df_perc[['BBRI', 'ICBP', 'SIDO', 'BMRI']].sum(axis=1)
vol_df_perc
```

```python
vol_df_perc_t = vol_df_perc.T
vol_df_ready = vol_df_perc_t.iloc[0:4, :]
vol_df_ready
```

```python
plt.figure(figsize=(16,10))
plt.stackplot(vol_df_ready.columns, vol_df_ready, labels=vol_df_ready.index,
              colors=['orange', 'green', 'magenta', 'lightblue'])
plt.legend()
plt.show()
```
